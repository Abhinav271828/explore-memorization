# Model Analysis
We analyze memorization for the first set of models. Config:

* Dataset: `Expr`, uniform PCFG; sizes $10^3, 10^4, 10^5$.
* Model: Transformer with `{'ntoken': 20, 'd_model': 64, 'nhead': 2, 'd_hid': 256, 'nlayers': 3}`
* Training: Adam with LR $10^{-3}$, early stopping with patience 3

```python
def get_model_name(size, epoch):
    epochs = {1000: 28, 10000: 7, 100000: 4} # The number of epochs each model was trained on.
    if epoch >= epochs[size]:
        raise ValueError("Epoch out of range")

    if epoch == 0:
        return f'../models/size={size}.0.ckpt.ckpt'
    return f'../models/size={size}.0.ckpt-v{epoch}.ckpt'
```
